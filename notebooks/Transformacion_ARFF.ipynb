Notebook para convertir los documentos .arff en .txt

```python
#Importamos las librerías necesarias
import pandas as pd
from scipy.io import arff #para poder abrir el código
```

```python
# Abrimos el archivo
arff_file = arff.loadarff('/content/GCM.arff')
df = pd.DataFrame(arff_file[0])
df
```

```python
# Los demás datos son numéricos, solo me interesa transformar una columna que es la del ground truth en números
df['class'].unique().tolist()
```

```python
# Lista de clases
clases = [
    'Breast', 'Prostate', 'Lung', 'Colorectal', 'Lymphoma', 'Bladder', 'Melanoma',
    'Uterus__Adeno', 'Leukemia', 'Renal', 'Pancreas', 'Ovary', 'Mesothelioma', 'CNS'
]

df['class'] = df['class'].str.decode('utf-8') # Decodifica valores byte a string

class_mapping = {nombre: i for i, nombre in enumerate(clases)} # mapeo
df['class'] = df['class'].map(class_mapping)  # Reemplaza directamente la columna el mapeo

df.to_csv("GCM.txt", sep=",", index=False, header=False) #Exporta el archivo en txt, separado en comas y sin el header e índices
```
